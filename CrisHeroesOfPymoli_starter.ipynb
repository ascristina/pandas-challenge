```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "/Users/acristina/opt/anaconda3/envs/PythonData/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data["SN"].count()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
summary_df = purchase_data.agg({"Item Name" : ['nunique', 'count'], "Price" : ['mean' , 'sum']})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
demo_df = purchase_data.groupby(["Gender"]).count()
df = demo_df.rename(index=str, columns={"Purchase ID": "Total Count"})
drop_df = df.drop(columns = ["SN", "Age", "Item ID", "Item Name", "Price"])
drop_df["Percentage of Players"] = df["Total Count"]/ (df["Total Count"].sum()) *100
drop_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
calc_df = purchase_data.groupby(['Gender'])
average_df = calc_df["Price"].mean()
average_df
add_df = calc_df["Price"].sum()
total_df = calc_df["Price"].count()

summary_df = pd.DataFrame({"Total Purchase Value": add_df,
                                    "Purchase Count": total_df,
                         "Average Purchase Price" : average_df})
summary_df["Normalized Purchasing"] = summary_df["Total Purchase Value"] / summary_df["Purchase Count"]
summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_range = purchase_data
age_range["Age Range"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
age_group = age_range.groupby("Age Range").count()
age_group = age_group.drop(columns = ["SN", "Age", "Item ID", "Item Name", "Price","Purchase ID"])
age_group = age_group.rename(index = str , columns = {'Gender' : 'Total Count'})
age_group['Percentage of Players'] = age_group['Total Count'] / (age_group['Total Count'].sum()) * 100
age_group
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_calc = age_range.groupby(['Age Range'])
age_add = age_calc['Price'].sum()
age_total = age_calc['Price'].count()
age_avg = age_calc['Price'].mean()
summary_age = pd.DataFrame({"Total Purchase Value": age_add,
                                    "Purchase Count": age_total,
                           "Average Purchase Price" : age_avg})
summary_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_group = purchase_data.groupby("SN")
sn_add = sn_group['Price'].sum()
sn_total = sn_group['Price'].count()
sn_summary = pd.DataFrame({"Total Purchase Value": sn_add,
                                    "Purchase Count": sn_total})
sn_summary["Average Purchase Price"] = sn_summary["Total Purchase Value"] / sn_summary["Purchase Count"]

sort_df = sn_summary.sort_values(["Total Purchase Value"], ascending = False)
sort_df.reset_index(inplace = True)
sort_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data.groupby(["Item Name", "Item ID",])
price_df = popular_df["Price"]
total_price = popular_df['Price'].sum()
count_price = popular_df['Price'].count()
item_summary = pd.DataFrame({"Total Purchase Value": total_price,
                                    "Purchase Count": count_price,
                            "Item Price": price_df})
item_summary.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profit_df = item_summary.sort_values(["Total Purchase Value"], ascending = False)
profit_df.reset_index(inplace = True)
profit_df.head()
```
